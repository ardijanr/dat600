```python
import matplotlib.pyplot as plt
import math
import time
```

```python
def read_runtime_file(filename):
    content = [[], []]
    with open(filename) as f:
        for line in f.readlines():
            split_line = line.split(",")
            if len(split_line) <= 1:
                continue
            content[0].append(int(split_line[0].strip()))
            content[1].append(int(split_line[1].strip()))
    return content

insertion_runtime_go = read_runtime_file("./go/runtime/insertionsort.txt")
heapsort_runtime_go = read_runtime_file("./go/runtime/heapsort.txt")
mergesort_runtime_go = read_runtime_file("./go/runtime/mergesort.txt")
quicksort_runtime_go = read_runtime_file("./go/runtime/quicksort.txt")

insertion_runtime_rust = read_runtime_file("./rust/runtime/insertionsort.txt")
heapsort_runtime_rust = read_runtime_file("./rust/runtime/heapsort.txt")
mergesort_runtime_rust = read_runtime_file("./rust/runtime/mergesort.txt")
quicksort_runtime_rust = read_runtime_file("./rust/runtime/quicksort.txt")

plt.plot(quicksort_runtime_go[0], quicksort_runtime_go[1], label="Go")
plt.plot(quicksort_runtime_rust[0], quicksort_runtime_rust[1], label="Rust")
plt.autoscale(enable=True, axis='y')
plt.title("Quicksort")
plt.legend()
plt.show()

plt.plot(mergesort_runtime_go[0], mergesort_runtime_go[1], label="Go")
plt.plot(mergesort_runtime_rust[0], mergesort_runtime_rust[1], label="Rust")
plt.autoscale(enable=True, axis='y')
plt.title("Merge Sort")
plt.legend()
plt.show()

plt.plot(insertion_runtime_go[0], insertion_runtime_go[1], label="Go")
plt.plot(insertion_runtime_rust[0], insertion_runtime_rust[1], label="Rust")
plt.autoscale(enable=True, axis='y')
plt.title("Insertion Sort")
plt.legend()
plt.show()

plt.plot(heapsort_runtime_go[0], heapsort_runtime_go[1], label="Go")
plt.plot(heapsort_runtime_rust[0], heapsort_runtime_rust[1], label="Rust")
plt.autoscale(enable=True, axis='y')
plt.title("Heap Sort")
plt.legend()
plt.show()
```
