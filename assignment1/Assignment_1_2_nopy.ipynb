```python
import matplotlib.pyplot as plt
import math
import time
```

```python
def read_runtime_file(filename):
    content = [[], []]
    with open(filename) as f:
        for line in f.readlines():
            split_line = line.split(",")
            if len(split_line) <= 1:
                continue
            content[0].append(int(split_line[0].strip()))
            content[1].append(int(split_line[1].strip()))
    return content



insertion_runtime_go = read_runtime_file("./go/runtime/insertionsort.txt")
heapsort_runtime_go = read_runtime_file("./go/runtime/heapsort.txt")
mergesort_runtime_go = read_runtime_file("./go/runtime/mergesort.txt")
quicksort_runtime_go = read_runtime_file("./go/runtime/quicksort.txt")

insertion_runtime_rust = read_runtime_file("./rust/runtime/insertionsort.txt")
heapsort_runtime_rust = read_runtime_file("./rust/runtime/heapsort.txt")
mergesort_runtime_rust = read_runtime_file("./rust/runtime/mergesort.txt")
quicksort_runtime_rust = read_runtime_file("./rust/runtime/quicksort.txt")


fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot for Quicksort
axs[0, 0].plot(quicksort_runtime_go[0], quicksort_runtime_go[1], label="Go")
axs[0, 0].plot(quicksort_runtime_rust[0], quicksort_runtime_rust[1], label="Rust")
axs[0, 0].autoscale(enable=True, axis='y')
axs[0, 0].set_title("Quicksort")
axs[0, 0].legend()

# Plot for Merge Sort
axs[0, 1].plot(mergesort_runtime_go[0], mergesort_runtime_go[1], label="Go")
axs[0, 1].plot(mergesort_runtime_rust[0], mergesort_runtime_rust[1], label="Rust")
axs[0, 1].autoscale(enable=True, axis='y')
axs[0, 1].set_title("Merge Sort")
axs[0, 1].legend()

# Plot for Insertion Sort
axs[1, 0].plot(insertion_runtime_go[0], insertion_runtime_go[1], label="Go")
axs[1, 0].plot(insertion_runtime_rust[0], insertion_runtime_rust[1], label="Rust")
axs[1, 0].autoscale(enable=True, axis='y')
axs[1, 0].set_title("Insertion Sort")
axs[1, 0].legend()

# Plot for Heap Sort
axs[1, 1].plot(heapsort_runtime_go[0], heapsort_runtime_go[1], label="Go")
axs[1, 1].plot(heapsort_runtime_rust[0], heapsort_runtime_rust[1], label="Rust")
axs[1, 1].autoscale(enable=True, axis='y')
axs[1, 1].set_title("Heap Sort")
axs[1, 1].legend()

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
